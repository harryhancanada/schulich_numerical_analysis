np.dot и np.inner идентичны для 1-мерных массивов, так что, вероятно, именно поэтому вы не замечаете никаких различий. Для n-мерных массивов они соответствуют общим тензорным операциям.

np.inner иногда называют "vector product" между тензором высшего и низшего порядка, особенно тензором, умноженным на вектор, и часто приводит к "tensor contraction". Она включает в себя умножение матрицы на вектор.

np.dot соответствует "тензорному произведению" и включает случай, упомянутый в нижней части страницы Википедии. Он обычно используется для умножения двух одинаковых тензоров, чтобы получить новый тензор. Она включает в себя матрицу-матричное умножение. 

```
a=np.array([[1,2],[3,4]])
b=np.array([[11,12],[13,14]])
```

```
np.dot(a,b)
```

```
np.dot(a.T, b)
```

```
np.dot(a, b.T)
```

```
np.inner(a,b)
```

```
np.outer(a, b)
```
