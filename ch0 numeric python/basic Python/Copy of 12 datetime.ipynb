```
import time
simpletime = time.time()
timetuple = time.localtime(simpletime)
print ("simpletime =", simpletime)
print ("timetuple =", timetuple)
print ("type(simpletime) =", type(simpletime))
print ("type(timetuple) =", type(timetuple))
```

```
import time
timestring = time.asctime( time.localtime(time.time()) )
print ("timestring =", timestring)
print ("type(timestring) =", type(timestring))
```

```
# From the datasets, we usually read strings. 
# Then we need to convert them to a date and time format. 
# Example 
from datetime import datetime 
datetime_str = '09/19/18 13:55:26' 
datetime_object = datetime.strptime(datetime_str, '%m/%d/%y %H:%M:%S') 
print(type(datetime_object)) 
print(datetime_object) # printed in default format
```
