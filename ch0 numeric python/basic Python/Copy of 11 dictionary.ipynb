```
dict1 = {'Name': 'Zara', 'Age': 7, 'Class': 'First'}
print(dict1.values()) # get all values
print(dict1.keys()) # get all keys
print(dict1.items()) # get all element of the dict
newdict = dict1.copy() # copy dict
print(newdict)
newdict.clear() # Removes all the elements from the dictionary
print("dict1 =", dict1)
print("newdict =", newdict)
# This did not change the original dictionary
```
