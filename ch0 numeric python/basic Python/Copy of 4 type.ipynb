```
# pipe assignment
a = b = c = 300  

# collective printing
print(a, b, c) 
```

```
# type change
var = 23.5
print(var)
var = "Now I'm a string"
print(var)
```

```
# printing not assigned value
print(300)
```

```
# first assignment
n = 300
```

```
# chain assignment
m = n
```

```
# reassignment
m = 400
# n is still the same
n
```

```
# another reassignment
n = "foo"
# At this point, constant 400 is orphaned, not referred by any variable
```
