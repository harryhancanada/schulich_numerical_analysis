```
from scipy.optimize import newton

def f(x):
    return (1.0/4.0)*x**3+(3.0/4.0)*x**2-(3.0/2.0)*x-2

x = 4

x0 = newton(f, x, fprime=None, args=(), tol=1.48e-08, maxiter=50, fprime2=None)

print('x: ', x)
print('x0: ', x0)
print("f(x0) = ", ((1.0/4.0)*x0**3+(3.0/4.0)*x0**2-(3.0/2.0)*x0-2 ))
```

```
import autograd.numpy as np
import autograd


def newton(f, x0, tol=1.48e-08, maxiter=50):
    g = autograd.grad(f)
    print("g =", g)
    print("g(np.array([3., 2.])) =", g(np.array([3., 2.])))
    h = autograd.hessian(f)
    print("h =", h)

    x = x0
    for _ in range(maxiter):
        delta = np.linalg.solve(h(x), -g(x))
        x = x + delta
        if np.linalg.norm(delta) < tol:
            break

    return x


def main():

    def f(x):
        return (x[0] ** 3) + (x[1] ** 3) - (9 * x[0] * x[1]) + 27

    x0 = np.array([3., 2.])

    print(newton(f, x0))


if __name__ == '__main__':
    main()
```
