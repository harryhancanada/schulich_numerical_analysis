Summing up values in a list can get tricky if the values are floating point numbers; digit cancellation can occur and the result may come out wrong. A classical example is the sum

```
1.0e16 + 1.0 - 1.0e16
```

The actual result is 1.0, but in double precision, this will result in 0.0. While in this example the failure is quite obvious, it can get a lot more tricky than that.
